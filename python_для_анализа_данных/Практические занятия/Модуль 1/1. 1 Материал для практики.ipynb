# Множественное присваивание

Множественное присваивание в Python позволяет обменивать значения двух переменных одной строкой кода.

```python
a = 10
b = 20

# Обмен значений переменных
a, b = b, a

print("a =", a)
print("b =", b)
```

# Условный оператор

```python
# Условный оператор

for i in range(6):
    if i % 2 == 0:
        print(f"i = {i}, четное число")
    else:
        print(f"i = {i}, нечетное число")
```

```python
# Несколько условий оформляются с помощью логических операторов and, or и not
x = int(input())
if x > 1 and x < 100:
    # Действия для числа в диапазоне (1; 100)
    print(x+50)
```

# Циклы for и while

```python
sequence = [1, 2, None, 4, None, 5]
total = 0
for value in sequence:
    if value is None:
        continue
    total += value
total
```

Функция range порождает последовательность равноотстоящих целых чисел:

```python
for i in range(0, 10, 2):
    print(i)
```

```python
for i in range(4):
    for j in range(4):
        if j > i:
            break
        print((i, j))
```

```python
x = 256
total = 0
iterat = 0
while x > 0:
    if total > 500:
        break
    total += x
    x = x // 2
    iterat += 1
print(total, iterat)
```

```python
#Методы строк

capitalize(): Возвращает копию строки, в которой первый символ имеет верхний регистр, а все остальные символы имеют нижний регистр.
swapcase(): Возвращает копию строки, в которой все символы, имеющие верхний регистр, преобразуются в символы нижнего регистра и наоборот.
title(): Возвращает копию строки, в которой первый символ каждого слова переводится в верхний регистр.
lower(): Возвращает копию строки, в которой все символы имеют нижний регистр.
upper(): Возвращает копию строки, в которой все символы имеют верхний регистр.

count(): Возвращает количество вхождений подстроки в строку.
startswith(): Проверяет, начинается ли строка с указанной подстроки.
endswith(): Проверяет, заканчивается ли строка указанной подстрокой.
find(): Возвращает индекс первого вхождения подстроки в строку.
rfind(): Возвращает индекс последнего вхождения подстроки в строку.
index(): Возвращает индекс первого вхождения подстроки в строку, начиная с указанного индекса.
rindex(): Возвращает индекс последнего вхождения подстроки в строку, начиная с указанного индекса.
strip(): Удаляет указанные символы с начала и конца строки.
lstrip(): Удаляет указанные символы с начала строки.
rstrip(): Удаляет указанные символы с конца строки.
replace(): Заменяет все вхождения указанной подстроки на другую подстроку.

isalnum(): Возвращает True, если все символы строки являются буквенными или цифровыми, и False в противном случае.
isalpha(): Возвращает True, если все символы строки являются буквенными, и False в противном случае.
isdigit(): Возвращает True, если все символы строки являются цифрами, и False в противном случае.
islower(): Возвращает True, если все символы строки находятся в нижнем регистре, и False в противном случае.
isupper(): Возвращает True, если все символы строки находятся в верхнем регистре, и False в противном случае.
isspace(): Возвращает True, если все символы строки являются пробелами (пробел, табуляция, новая строка и т.д.), и False в противном случае.
```

```python
# Метод format()

# Передаем необходимые параметры методу format(), а Python ставит их вместо фигурных скобок {} – заполнителей
# Мы можем создавать сколько угодно заполнителей

birth_year = 1980
name = 'Alex'
profession = 'engeener'
text = 'My name is {}, I was born in {}, I work as a {}.'.format(name, birth_year, profession)

print(text)
```

```python
# Для наглядности и гибкости форматирования мы можем использовать порядковый номер в заполнителе: {0}, {1}, {2} и т.д.

# Такой номер определяет позицию параметра, переданного методу format() (нумерация начинается с нуля)

birth_year = 1980
name = 'Alex'
profession = 'engeener'
text = 'My name is {0}, I was born in {1}, I work as a {2}.'.format(name, birth_year, profession)

print(text)
```

В Python есть f-строки (сокращение от «форматированные строковые литералы»), благодаря которым создание форматированных строк стало очень удобным. Чтобы создать строку, поместите букву f непосредственно перед строковым литералом. А внутри строки заключайте выражения Python в фигурные скобки, тогда вместо них в строку будет подставлено значение выражения:

```python
amount = 10
rate = 95.46
currency = "Rubles"
result = f"{amount} {currency} is worth US${amount / rate}"
result
```

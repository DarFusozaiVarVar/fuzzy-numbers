```python
import matplotlib.pyplot as plt
import numpy as np

y_positions = np.arange(3) # Positions for horizontal bars
start_values = np.array([1, 3, 6])
end_values = np.array([4, 7, 8])
bar_labels = ['Interval 1', 'Interval 2', 'Interval 3']

# Calculate width and left position for barh
widths = end_values - start_values
left_positions = start_values

plt.barh(y_positions, widths, left=left_positions, height=0.5, align='center', color='lightgreen', edgecolor='black')
plt.yticks(y_positions, bar_labels) # Set y-axis labels
plt.xlabel('Value Range')
plt.ylabel('Interval')
plt.title('Horizontal Bars Representing Intervals')
plt.show()
```
